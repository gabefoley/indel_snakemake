```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("indel_output.csv")
```

```python
sorted_df = df.sort_values('Root_Dist')
sorted_df.head(50)
```

```python
methods = ['PSP', 'PSML', 'SICP', 'SICML', 'BEP', 'BEML', 'fastmlp', 'fastmlml']
categories = ['insertion', 'deletion']
df

taxa = [x for x in df['Taxa'].unique()]

# df1 = df[(df['category'] == 'A') & (df['value'].between(10,20))]

# subset_df = df[['Name', 'Taxa'] + methods]
# df.loc[df["col_z"] < m, ["col_x", "col_y"]]

for taxon in taxa:
    print (taxon)
    for category in categories:
        print (category)

        taxon_df = sorted_df.loc[(df['Taxa'] == taxon) & (df['Category'] == category), ['Name', 'Root_Dist'] + methods]
        fig,ax = plt.subplots(1,1,figsize=(20,2))

        sns.barplot(x="Name", y="Root_Dist", data=taxon_df, ax=ax)

        taxon_df = sorted_df.loc[(df['Taxa'] == taxon) & (df['Category'] == category), ['Name'] + methods]


        taxon_df = taxon_df.set_index('Name')

        print(taxon_df)
        for method in methods:
            print(taxon_df[method].sum())

        fig,ax = plt.subplots(1,1,figsize=(20,10))
        my_colors=['red','lightgreen']

        sns.heatmap(taxon_df.transpose(), cmap=my_colors, cbar=False, square=True, linewidth=0.1, linecolor=(0.2,0.2,0.2), ax=ax)

        colorbar = ax.collections[0].colorbar
    #     M=dt_tweet_cnt.max().max()
    #     colorbar.set_ticks([1, 2])
    #     colorbar.set_ticklabels(['low','med'])
        plt.show()
    #     taxon_df = df[(df['Name'] + methods) & (df['Taxa'] == taxon)]
```
